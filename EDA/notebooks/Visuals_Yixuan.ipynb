```python
import librosa
import librosa.display
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
file = "../the-circor-digiscope-phonocardiogram-dataset-1.0.3/training_data/2530_AV.wav"
```

```python
signal, sr = librosa.load(file)
```

```python
plt.figure(figsize=(10, 4))
librosa.display.waveshow(signal, sr=sr)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
fft = np.fft.fft(signal)
```

```python
magnitude = np.abs(fft)
```

```python
frequency = np.linspace(0,sr,len(magnitude))
```

```python
frequency
```

```python
len(frequency) == len(magnitude)
```

```python
left_frequency = frequency[:int(len(frequency)/2)]
left_magnitude = magnitude[:int(len(frequency)/2)]

plt.plot(left_frequency,left_magnitude)
plt.xlabel('frequency')
plt.ylabel('magnitude')
plt.show()
```

```python
D = librosa.stft(signal)  
amplitude = librosa.amplitude_to_db(abs(D), ref=np.max)

plt.figure(figsize=(10, 4))
librosa.display.specshow(amplitude, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.tight_layout()
plt.show()
```
